```python
# Solution 3
# Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

import math

def normal_pdf(x, mean, std_dev):
    """
    Calculates the PDF of a normal distribution at a given point.

    Args:
        x (float): The value at which to evaluate the PDF.
        mean (float): Mean (average) of the normal distribution.
        std_dev (float): Standard deviation of the normal distribution.

    Returns:
        float: The PDF value at the specified point.
    """
    if std_dev <= 0:
        raise ValueError("Standard deviation must be positive.")
    
    exponent = -0.5 * ((x - mean) / std_dev) ** 2
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    pdf_value = coefficient * math.exp(exponent)
    
    return pdf_value

# Example usage:
mean_value = 10.0
std_dev_value = 2.0
point_to_evaluate = 12.0
pdf_result = normal_pdf(point_to_evaluate, mean_value, std_dev_value)
print(f"PDF at x = {point_to_evaluate:.2f}: {pdf_result:.6f}")
```

```python
# Solution  4
# What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

# Properties of Binomial Distribution:
# The binomial distribution is characterized by the following properties:
# Two Possible Outcomes: It applies to situations with only two distinct outcomes, such as success/failure, yes/no, or true/false.
# Fixed Number of Trials: There is a fixed number of n independent trials or repeated experiments.
# Constant Probability: The probability of success or failure remains the same for each trial.
# Examples of Events Where Binomial Distribution Can Be Applied:
# Tossing a Coin:
# When you flip a fair coin, there are only two possible outcomes: heads or tails.
# The number of heads obtained in a series of coin tosses follows a binomial distribution.
# Quality Control in Manufacturing:
# Consider a factory producing light bulbs. Each bulb can be either defective or non-defective.
# The binomial distribution helps model the probability of finding a certain number of defective bulbs in a sample.
```

```python
# Solution 5
# Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.

import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_trials = 1000  # Number of trials (sample size)
p_success = 0.4  # Probability of success

# Generate the random sample
binomial_sample = np.random.binomial(n=n_trials, p=p_success, size=1000)

# Plot a histogram
plt.figure(figsize=(8, 6))
plt.hist(binomial_sample, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.title(f"Binomial Distribution (n={n_trials}, p={p_success})")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Solution 6
# Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.

import math

def poisson_cdf(x, mean_lambda):
    """
    Calculates the cumulative distribution function (CDF) of a Poisson distribution.

    Args:
        x (int): The point at which to evaluate the CDF.
        mean_lambda (float): The mean of the Poisson distribution.

    Returns:
        float: The probability that the random variable is less than or equal to x.
    """
    cdf = 0.0
    for k in range(x + 1):
        cdf += math.exp(-mean_lambda) * (mean_lambda ** k) / math.factorial(k)
    return cdf

# Example usage:
mean_error_rate = 7  # Mean number of errors per 5,000 lines of code
selected_errors = 2  # Number of errors we want to find the probability for
cdf_value = poisson_cdf(selected_errors, mean_error_rate)
print(f"The probability of exactly {selected_errors} errors in 5,000 lines of code is approximately {cdf_value:.4f}")
```

```python
# Solution 7
# How Binomial distribution different from Poisson distribution?

# Nature:
# Binomial Distribution:
# Deals with a fixed number of trials (e.g., flipping a coin 3 times).
# Only two possible outcomes: success or failure.
# Poisson Distribution:
# Describes the count of independent events occurring randomly within a given period of time.
# No limit on the number of possible outcomes.
# Parameters:
# Binomial Distribution:
# Biparametric: Characterized by two parameters: n (number of trials) and p (success probability).
# Mean and variance: Âµ = np, ÏƒÂ² = npq (where q = 1 - p).
# Poisson Distribution:
# Uniparametric: Depends on a single parameter: Î» (average rate of events per unit time or space).
# Mean and variance: Both equal to Î».
# Examples:
# Binomial Distribution:
# Coin tossing experiment (success = heads, failure = tails).
# Poisson Distribution:
# Counting printing mistakes per page in a large book.
```

```python
# Solution 8
# Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.

import numpy as np
sample_size = 1000
mean_lambda = 5
random_sample = np.random.poisson(mean_lambda, sample_size)
# Calculate sample mean
sample_mean = np.mean(random_sample)

# Calculate sample variance
sample_variance = np.var(random_sample, ddof=1)

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Variance: {sample_variance:.4f}")
```

```python
# Solution 9
# How mean and variance are related in Binomial distribution and Poisson distribution?

# Binomial Distribution:
# Mean (Î¼):
# The mean of a Binomial distribution is given by: [ \mu = n \cdot p ]
# Here, (n) represents the number of trials (or experiments), and (p) is the probability of success in each trial.
# Variance (ÏƒÂ²):
# The variance of a Binomial distribution is given by: [ \sigma^2 = n \cdot p \cdot (1 - p) ]
# Poisson Distribution:
# Mean and Variance:
# In the Poisson distribution, both the mean (average) and variance are equal and are denoted by the parameter (Î») (lambda).
# This property of equal mean and variance is a distinctive characteristic of the Poisson distribution and simplifies its statistical analysis.
```
