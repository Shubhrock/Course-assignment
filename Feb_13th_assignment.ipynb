```python
#solution 3
#What errors are defined in the ArithmeticError class? Explain any two with an example.

#Errors defined in ArithmeticError are:
#--> ZeroDivisionError
#--> OverflowError
#--> FloatingPointError

#OverflowError example

j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
    
#ZeroDivisionError example
try:
    1/0
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

```python
#Solution 4
#Why LookupError class is used? Explain with an example KeyError and IndexError.

#Answer
#LookupError Exception is the Base class for errors raised when something canâ€™t be found. The base class for the exceptions that 
#are raised when a key or index used on a mapping or sequence is invalid.

#IndexError
#An IndexError is raised when a sequence reference is out of range.
try:
    foo = ['a', 's', 'd', 'f', 'g']
    print (foo[5])
except IndexError as e:
    print (e)

    
#KeyError
#A KeyError is raised when a value is not found as a key of a dictionary.
try:
    s = {'a':5, 'b':7}['c']

except KeyError as e:
    print (e)
```
