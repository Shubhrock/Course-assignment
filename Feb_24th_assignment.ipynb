```python
#Solution 8

# You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects
# all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The
# function should return a new DataFrame that contains only the selected rows.
# For example, if df contains the following values:
# A B C
# 0 3 5 1
# 1 8 2 7
# 2 6 9 4
# 3 2 3 5
# 4 9 1 2

# Your function should select the following rows: A B C
# 1 8 2 7
# 2 6 9 4
# 4 9 1 2
# The function should return a new DataFrame that contains only the selected rows.

import pandas as pd

data2 = {
    "A" : [3,8,6,2,9],
    "B" : [5,2,9,3,1],
    "C" : [1,7,4,5,2]
}
df6 = pd.DataFrame(data2)
result = df6[(df6["A"] > 5) & (df6["B"] < 10)]
result
```

```python
#Solution 9

# Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean,
# median, and standard deviation of the values in the 'Values' column.

import pandas as pd

data = {
  "Values": [155, 220, 310, 189, 174]
}

df7 = pd.DataFrame(data)
mean_of_values = df7.mean()
print("Mean of", mean_of_values)

median_of_values = df7.median()
print("Median of", median_of_values)

median_of_values = df7.std()
print("Standard Deviation of", median_of_values)
```

```python
#Solution 10

# Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to
# create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days
# for each row in the DataFrame. The moving average should be calculated using a window of size 7 and
# should include the current day.

import pandas as pd

data = {
  "Sales": [2871, 2765.9, 3884.34, 3746.7, 5205.27, 3479.76, 2497.77, 5512.32, 2168.54, 4708.44],
  "Date" : ['29/10/2023','30/10/2023','31/10/2023','1/11/2023','2/11/2023','3/11/2023','4/11/2023','5/11/2023','6/11/2023','7/11/2023']
}

df8 = pd.DataFrame(data)
df8['MovingAverage'] = df8['Sales'].rolling(7).mean()
df8
```

```python
# Solution 11

# You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new
# column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g.
# Monday, Tuesday) corresponding to each date in the 'Date' column.

import datetime
import pandas as pd

data = {
  "Date" : [datetime.date(2023, 11, 7),
            datetime.date(2023, 11, 8),
            datetime.date(2023, 11, 9),
            datetime.date(2023, 11, 10),
            datetime.date(2023, 11, 11),
            datetime.date(2023, 11, 12),
            datetime.date(2023, 11, 13)           
           ]
}

df9 = pd.DataFrame(data)
df9["Weekday"] = df9["Date"].apply(lambda x: datetime.datetime.strftime(x, '%A'))
df9
```

```python
#Solution 12

# Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python
# function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

import datetime
import pandas as pd

data = {
  "Date" : [datetime.date(2023, 1, 17),
            datetime.date(2023, 1, 29),
            datetime.date(2023, 5, 19),
            datetime.date(2023, 1, 10),
            datetime.date(2023, 11, 11),
            datetime.date(2023, 4, 12),
            datetime.date(2023, 1, 23)           
           ]
}

df10 = pd.DataFrame(data)


# Select DataFrame rows between two dates
mask = (df10['Date'] > datetime.date(2023, 1, 1)) & (df10['Date'] <= datetime.date(2023, 1, 31))
df10 = df10.loc[mask]
df10
```
